과제 1. faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라

```python
from faker import Faker
import random
import pandas as pd

fake = Faker()

# 클래스 정의
class User:
    def __init__(self):
        self.name = fake.name()
        self.age = random.randint(20, 70)
        self.email = fake.email()
        self.job = fake.job()

        self.credit_amount = random.randint(1000, 20000)
        self.duration = random.randint(6, 60)

        # class 생성
        if self.credit_amount > 15000 or self.duration > 36:
            self.class_label = "bad"
        else:
            self.class_label = "good"

    def to_dict(self):
        return {
            "name": self.name,
            "age": self.age,
            "email": self.email,
            "job": self.job,
            "credit_amount": self.credit_amount,
            "duration": self.duration,
            "class": self.class_label
        }

# 1000명 생성
data = []

for _ in range(1000):
    user = User()
    data.append(user.to_dict())

df = pd.DataFrame(data)

df.head()
```

Faker를 사용하여 가상의 사용자 데이터를 생성하고,
class를 정의하여 각 데이터를 객체 형태로 구성하였다.
반복문을 통해 1000명의 데이터를 생성하고 DataFrame으로 변환하였다.

과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)

```python
class User:
    def __init__(self, name):
        # HAS (Attributes)
        self.name = name
        self.following = 0
        self.follower = 0

    # DOES (Method)
    def following_mode(self, other):
        print(f"{self.name} is following {other.name}")

        #속성 변경
        self.following += 1
        other.follower += 1


# 객체 생성
tom = User("Tom")
jerry = User("Jerry")

# 10번 following
for i in range(10):
    tom.following_mode(jerry)

# 결과 확인
print("\n--- Result ---")
print(f"Tom - following: {tom.following}, follower: {tom.follower}")
print(f"Jerry - following: {jerry.following}, follower: {jerry.follower}")
```
